We will now examine the behaviour of a so called *Gaussian* or *Normal* distribution in a 2D grid and how its form changes with multiple parameters.

The following formula describes the probability density function of a normal distribution in 2D space (x and y):
$$f(x,y)=\frac{1}{\sqrt{2\pi\sigma^2}}\operatorname{exp}\left\{-\frac{\left(x-\mu_{x}\right)^2+(y-\mu_{y})^2}{2\sigma^2}\right\}$$

* $\mu_{x}$ and $\mu_{y}$ describe the mean value in the x and y axis, which means that you can find the maximum of the distribution in that axis at their respective mean values.
* $\sigma$ describes the standard deviation, which as the name implies tells you how densely packed around the mean the distribution will be.

## Creating a function
In the following code section a function is defined that creates a Gaussian distribution, that's the function we discussed at the beginning of this document:
$$f(x,y)=\frac{1}{\sqrt{2\pi\sigma^2}}\operatorname{exp}\left\{-\frac{\left(x-\mu_{x}\right)^2+(y-\mu_{y})^2}{2\sigma^2}\right\}$$

- You can create a function which takes in several inputs and gives out several outputs at any point of your program. It is a good coding practice to create a function if you feel it makes your code more readable, or if you will be repeatedly computing the same function but with different values.
- Creating a function spares you from writing the same code lines multiple times.

**<font color=red>You DON'T have to change anything in the following box with code. Leave it as it is and use it as a reference for the next Jupyter Notebook of today's class!</font>**

````python
def gaussianDistribution(gridPoints, standardDeviation, amplitude, x_strike, y_strike, leftBoundary, rightBoundary):
    x = np.linspace(leftBoundary, rightBoundary, gridPoints)
    y = np.linspace(leftBoundary, rightBoundary, gridPoints)
    X, Y = np.meshgrid(x, y)

    z = amplitude*np.exp(standardDeviation*(np.power(X-x_strike,2)+np.power(Y-y_strike,2)))

    return z
````

The method **np.linspace** gives out a vector with *n* evenly-spaced numbers, between a left and a right boundary. So for example:
```python
    x = np.linspace(leftBoundary, rightBoundary, gridPoints)

```
Will give out a vector with exactly *gridPoints* number of elements, evenly spaced out between the two numbers *leftBoundary* < *rightBoundary*

We need to create such vectors, because we will later be solving problems in 2D space, and since we cannot analytically solve many problems, what we do is solve them at only specific points in space... points that we choose to place evenly throughout or grid.

```python
    X, Y = np.meshgrid(x, y)
```
**np.meshgrid()** creates a grid out of the two vectors we just created. If you do not know what a grid is [check the reference page for this method](https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html) or ask a tutor to explain you this concept.

***
*The method **np.power** computes the power (^) of every value of an array.

*The method **np.exp** computes the exponential value, also element-wise, for all the elements of a matrix or an array.

And finally the expression:
```python     
    z = amplitude*np.exp(standardDeviation*(np.power(X-x_strike,2)+np.power(Y-y_strike,2)))
```
will create a matrix z that solves the Gaussian distribution function for every point of the grids X,Y which you created before.

````python
leftBoundary = -3
rightBoundary = 6
standardDeviation = -0.6

x = np.linspace(leftBoundary, rightBoundary, gridPoints)
y = np.linspace(leftBoundary, rightBoundary, gridPoints)
X, Y = np.meshgrid(x, y)
````

## Exercise 3

The variables **x_strike** and **y_strike** represent the points in the grid where the drummer *strikes* the drum membrane with the sticks. The deformation of the drum membrane right after getting hit is being model here as a Gaussian distribution. Therefore, **x_strike** and **y_strike** represent our **mean values** (where the maximum is).

Please place the maximum of the distribution at the following points:

1. x coordinate of the maximum should be at 1/3 of whole length of the x-axis of the grid. y coordinate of the maximum should be at 1/2 of the whole length of the y-axis of the grid

2. at (0,0)

````python
x_strike = x[(math.floor(x.size/2))]
y_strike = y[(math.floor(y.size/2))]



z = gaussianDistribution(gridPoints,standardDeviation,amplitude,x_strike,y_strike,leftBoundary,rightBoundary)


````

## Creating a plot

The following code creates a 3D surface of the Gaussian probability density over the 2D grid created with the *meshgrid()* method of Numpy.
If you want to learn more about what the different parameters of the method *plot_surface* do, like *cmap* or *edgecolor*, check [this link!](https://matplotlib.org/mpl_toolkits/mplot3d/tutorial.html#surface-plots)

````python
fig = plt.figure() # initialization of a figure
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, z, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
````
